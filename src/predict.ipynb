# predict.ipynb

Single-image inference notebook for the fruit classification model.

## Notes

- Predict a fruit class from one image using the saved Keras model.

```python
from pathlib import Path
import numpy as np
from PIL import Image
import tensorflow as tf

CLASS_NAMES = [
    "Apple Granny Smith", "Apricot", "Avocado", "Banana", "Blueberry",
    "Cactus fruit", "Cherry", "Corn", "Kiwi", "Mango",
    "Orange", "Pineapple", "Strawberry", "Watermelon",
]

def load_and_prepare_image(image_path: Path, img_size: int = 64) -> np.ndarray:
    image = Image.open(image_path).convert("RGB")
    image = image.resize((img_size, img_size))
    array = np.asarray(image, dtype=np.float32) / 255.0
    return np.expand_dims(array, axis=0)

def predict_one(
    image_path,
    model_path="../model/fruit_classifier_final_best_model.keras",
    img_size=64,  # <- match model input
):
    model = tf.keras.models.load_model(model_path)
    image_array = load_and_prepare_image(Path(image_path), img_size)

    probs = model.predict(image_array, verbose=0)[0]
    pred_idx = int(np.argmax(probs))

    print("Predicted class :", CLASS_NAMES[pred_idx])
    print("Confidence      :", f"{float(probs[pred_idx]):.4f}")
    print("\nTop-3 predictions:")

    top3_idx = np.argsort(probs)[::-1][:3]
    for idx in top3_idx:
        print(f"- {CLASS_NAMES[int(idx)]}: {float(probs[idx]):.4f}")

predict_one("test_images/pineapple.png")
```
